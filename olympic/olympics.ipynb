```python
import pandas as pd
import numpy as np
```

```python
def load_data():
    df = pd.read_csv('data\olympics.csv',skiprows=[0])
    df.columns = ['country','#summer', 'Gold', 'Silver','Bronze','Total','#Winter', 'Gold', 'Silver','Bronze','Total','#Game', 'Gold', 'Silver','Bronze','Combined Total']
    df.replace(u'\xa0',u'', regex=True, inplace=True)
    df = df.replace('\(.+?\)|\[.+?\]\s*','',regex=True).set_index('country')
    
    df=df.drop('Total',axis=1)
    df=df.drop('Combined Total',axis=1)
    return df

    
    
```

```python
df=load_data()
df

```

```python
def gold_medal(df):
    a=df.values
    n=a.shape
    a=np.delete(a, n[0]-1, 0)
    x=a[:, 1]
    max=a.max(axis=0)
    max=max[1]
    

    index = np.where(x==max)

    country=df.index.values
    for i in index:
        temp=country[i]
    str1 = " ".join(temp)
    return str1.strip()
    
    
    
```

```python
gold_medal(df)
```

```python
def biggest_difference_in_gold_medal(df):
    a=df.values
    n=a.shape
    a=np.delete(a, n[0]-1, 0)
    x=a[:, 1]
    y=a[:, 5]
    z=np.subtract(x, y)
    max=z.max(axis=0)
    index = np.where(z==max)
    country=df.index.values
    for i in index:
        temp=country[i]
    str1 = " ".join(temp)
    return str1.strip()

    
```

```python
biggest_difference_in_gold_medal(df)
```

```python
def get_points(df):
    a=df.values
    n=a.shape
    x=a[:, 9]
    y=a[:, 10]
    z=a[:,11]
    x=np.multiply(x, 3)
    y=np.multiply(y, 2)
    df=df.assign(Points = x + y + z)
    x=df['Points']
    return x
    
```
